Revoir l’exercice 5, en utilisant les listes et les fonctions pour saisir des nombres réels et pour calculer les fonctions suivantes (sans utiliser Numpy) :
la somme
la moyenne
le produit
le max
le min
l’écart-type
la variance
la covariance
Pour effectuer cet exercice, vous aurez besoin d'utiliser la fonction sqrt pour l'utilisation de la fonction racine carée
Pour cela, ajoutez en haut de l'exercice la ligne suivante :
from math import sqrt

```python
from math import sqrt
def saisir_nombres():
    nombres = []
    while True:
        entree = input("Entrez un nombre réel (ou 'fin' pour terminer) : ")
        if entree.lower() == 'fin':
            break
        try:
            nombres.append(float(entree))
        except ValueError:
            print("Veuillez entrer un nombre valide.")
    return nombres

def somme(nombres):
    return sum(nombres)
def moyenne(nombres):
    return somme(nombres) / len(nombres) if nombres else 0
def produit(nombres):
    result = 1
    for n in nombres:
        result *= n
    return result
def maximum(nombres):
    return max(nombres) if nombres else None
def minimum(nombres):
    return min(nombres) if nombres else None
def variance(nombres):
    m = moyenne(nombres)
    return sum((x - m) ** 2 for x in nombres) / len(nombres) if nombres else 0
def ecart_type(nombres):
    return sqrt(variance(nombres))
def covariance(nombres1, nombres2):
    if len(nombres1) != len(nombres2) or not nombres1:
        return None
    m1 = moyenne(nombres1)
    m2 = moyenne(nombres2)
    return sum((nombres1[i] - m1) * (nombres2[i] - m2) for i in range(len(nombres1))) / len(nombres1)


nombres = saisir_nombres()

if nombres:
    print("Somme :", somme(nombres))
    print("Moyenne :", moyenne(nombres))
    print("Produit :", produit(nombres))
    print("Max :", maximum(nombres))
    print("Min :", minimum(nombres))
    print("Écart-type :", ecart_type(nombres))
    print("Variance :", variance(nombres))
    nombres2 = saisir_nombres()
    if nombres2:
        cov = covariance(nombres, nombres2)
        if cov is not None:
            print("Covariance :", cov)
        else:
            print("Les listes doivent avoir la même longueur pour calculer la covariance.")
else:
    print("Aucun nombre n'a été saisi.")
```

```python
liste_personnes = []
def personne():
    N = input("Quel est votre nom ? ")
    P = input("Quel est votre prénom ? ") 
    G = input("Quel est votre genre ? ") 
    A = int(input("Quel est votre âge? ")) 
    return {"nom": N, "prenom": P, "genre": G, "age": A}

def ajouter_personne(liste):
    nouvelle_personne = personne()
    liste.append(nouvelle_personne)
def lister_personnes(liste):
    if not liste:
        print("La liste est vide.")
    else:
        print("Liste des personnes :")
        for i, p in enumerate(liste):
            print(f"{i + 1}. Nom: {p['nom']}, Prénom: {p['prenom']}, Genre: {p['genre']}, Âge: {p['age']}")
def chercher_personne(liste, nom):
    for p in liste:
        if p['nom'].lower() == nom.lower():
            print(f"Personne trouvée : Nom: {p['nom']}, Prénom: {p['prenom']}, Genre: {p['genre']}, Âge: {p['age']}")
            return
    print("Personne non trouvée.")
if __name__ == "__main__":
    while True:
        print("\nMenu:")
        print("1. Ajouter une personne")
        print("2. Lister les personnes")
        print("3. Chercher une personne")
        print("4. Quitter")
        choix = input("Choisissez une option (1-4) : ")

        if choix == '1':
            ajouter_personne(liste_personnes)
        elif choix == '2':
            lister_personnes(liste_personnes)
        elif choix == '3':
            nom_a_chercher = input("Entrez le nom de la personne à chercher : ")
            chercher_personne(liste_personnes, nom_a_chercher)
        elif choix == '4':
            print("Au revoir!")
            break
        else:
            print("Choix invalide. Veuillez réessayer.")
```
